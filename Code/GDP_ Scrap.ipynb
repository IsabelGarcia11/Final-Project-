```python
url='https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'
html=req.get(url).content
html[:10]
```

```python
soup=bs(html, 'html.parser')
type(soup)
```

```python
tabla=soup.find_all('table')[2]
type(tabla)
```

```python
filas=tabla.find_all('tr')
filas=[f.text.strip().split('\n') for f in filas][2:]
filas[:2]
```

```python
final=[]
for e in filas:
    tmp=[]
    for st in e:
        if st!='':
            tmp.append(st)
    final.append(tmp)   
final[0]
```

```python
col=['country','region','gdp 2021 (us$ million)','Year_','GDP (UN)','Year1','GDP (WB)','Year2']
```

```python
#df
data=final[0:]  #no coger las celdas combinadas
df=pd.DataFrame(data, columns=col)
df.head(10)
```

```python
df=df.drop(columns=['Year_','GDP (UN)','Year1','GDP (WB)','Year2'])
df
```
